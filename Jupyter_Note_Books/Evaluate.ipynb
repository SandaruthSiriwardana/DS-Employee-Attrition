```python
import pandas as pd

# Assuming you have a DataFrame called 'salary' with 'Net_salary' and 'Employee_No' columns

# Group by 'Employee_No' and calculate the average net salary
average_salary = salary.groupby('Employee_No')['Net_salary'].mean().reset_index()

# Rename the column
average_salary = average_salary.rename(columns={'Net_salary': 'Average_Net_Salary'})

# Preview the new DataFrame
# print(average_salary)
```

```python
# Merge the DataFrames based on the common 'Employee_No' column
merged_df = pd.merge(average_salary, employee[['Employee_No', 'Status']], on='Employee_No', how='left')

merged_df
```

```python
# find null
merged_df.isnull().sum()
# remove null
merged_df = merged_df.dropna()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'merged_df' with 'Employee_No', 'Average_Net_Salary', and 'Status' columns

# Plotting the scatter plot
plt.scatter(merged_df['Average_Net_Salary'], merged_df['Status'])

# Adding labels and title to the plot
plt.xlabel('Average Net Salary')
plt.ylabel('Employee Status')
plt.title('Relationship between Average Net Salary and Employee Status')

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'merged_df' with 'Employee_No', 'Average_Net_Salary', and 'Status' columns

# Create a box plot using seaborn
sns.boxplot(x='Status', y='Average_Net_Salary', data=merged_df)

# Calculate the five-number summary
summary = merged_df.groupby('Status')['Average_Net_Salary'].describe()[['min', '25%', '50%', '75%', 'max']]

# Print the five-number summary
print(summary)

# Add labels and title to the plot
plt.xlabel('Employee Status')
plt.ylabel('Average Net Salary')
plt.title('Relationship between Average Net Salary and Employee Status')

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'merged_df' with 'Employee_No', 'Average_Net_Salary', and 'Status' columns
# Calculate the five-number summary
summary = merged_df.groupby('Status')['Average_Net_Salary'].describe()[['min', '25%', '50%', '75%', 'max']]

# Print the five-number summary
print(summary)
# Calculate the interquartile range (IQR) for 'Average_Net_Salary'
Q1 = merged_df['Average_Net_Salary'].quantile(0.25)
Q3 = merged_df['Average_Net_Salary'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
filtered_df = merged_df[(merged_df['Average_Net_Salary'] >= lower_bound) & (merged_df['Average_Net_Salary'] <= upper_bound)]

# Create a box plot of the filtered DataFrame
sns.boxplot(x='Status', y='Average_Net_Salary', data=filtered_df)

# Add labels and title to the plot
plt.xlabel('Employee Status')
plt.ylabel('Average Net Salary')
plt.title('Relationship between Average Net Salary and Employee Status ')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'employee' with 'Inactive_year' and 'Status' columns

# Filter the DataFrame for the desired Inactive_years (2020, 2021, 2022, and 2023)
Inactive_years = [2020, 2021, 2022, 2023]
filtered_data = employee[employee['Inactive_year'].isin(Inactive_years)]

# Count the frequency of each status for each Inactive_year
grouped_data = filtered_data.groupby(['Inactive_year', 'Status']).size().unstack()

# Create the bar plot
grouped_data.plot(kind='bar', stacked=True)

# Add labels and title to the plot
plt.xlabel('Status')
plt.ylabel('Frequency')
plt.title('Frequency of Status for  Resigned and inactive 2020-2023')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'employee' with 'Experience' and 'Status' columns

# Group the data by 'Experience' and 'Status' and count the frequency
grouped_data = employee.groupby(['Experience', 'Status']).size().unstack()

# Create the clustered bar chart
grouped_data.plot(kind='bar', stacked=False)

# Add labels and title to the plot
plt.xlabel('Working Experience')
plt.ylabel('Frequency')
plt.title('Working Experience and Status')

# Add legend
plt.legend(title='Status')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame called 'employee'' and 'Status' columns

# Define the working experience categories
experience_categories = range(0, 10)

# Define the employment status categories
status_categories = ['Active', 'Inactive', 'Resigned']

# Create an empty DataFrame to store the frequencies
frequencies = pd.DataFrame(0, index=experience_categories, columns=status_categories)

# Calculate the frequencies of each working experience and status combination
for i, exp in enumerate(experience_categories):
    for status in status_categories:
        frequencies.loc[exp, status] = len(employee[(employee['Experience'] == exp) & (employee['Status'] == status)])

# Create the clustered bar chart
frequencies.plot(kind='bar', stacked=False, width=0.8)

# Add labels and title to the plot
plt.xlabel('Working Experience')
plt.ylabel('Frequency')
plt.title('Working Experience vs. Employment Status')

# Customize the x-axis tick labels
plt.xticks(np.arange(len(experience_categories)), experience_categories, rotation=45)

# Add legend
plt.legend(title='Status')

# Display the plot
plt.show()
```
