```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'employee' with 'Employee_No', 'Status', 'Marital_Status', and 'Religion_ID' columns

# Create a cross-tabulation of 'Status', 'Marital_Status', and 'Religion_ID'
cross_tab = pd.crosstab([employee['Status'], employee['Marital_Status']], employee['Religion_ID'])

# Create subplots for each state
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# List of states
states = employee['Status'].unique()

# Iterate over each state
for i, state in enumerate(states):
    # Get the cross-tabulation for the current state
    state_cross_tab = cross_tab.loc[state]
    
    # Calculate the total count for each Religion_ID
    religion_counts = state_cross_tab.sum()
    
    # Plot the pie chart for the current state
    ax = axes[i]
    ax.pie(religion_counts, labels=religion_counts.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Religion_ID Distribution for {state}')
    ax.axis('equal')

# Display the pie charts
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame called 'employee' with 'Experience' and 'Status' columns

# Define the working experience categories
experience_categories = range(0, 31)

# Define the employment status categories
status_categories = ['Active', 'Inactive', 'Resigned']

# Create an empty DataFrame to store the frequencies
frequencies = pd.DataFrame(0, index=experience_categories, columns=status_categories)

# Calculate the frequencies of each working experience and status combination
for i, exp in enumerate(experience_categories):
    for status in status_categories:
        frequencies.loc[exp, status] = len(employee[(employee['Experience'] == exp) & (employee['Status'] == status)])

# Figure 1: Range 0-31 with a long x-axis
fig1, ax1 = plt.subplots()

# Smooth each line using a rolling window average with window size 3
smoothed_data = frequencies.rolling(window=3, center=True).mean()
smoothed_data.plot(kind='line', ax=ax1, linestyle='--', linewidth=2)

# Add labels and title to the plot
ax1.set_xlabel('Working Experience')
ax1.set_ylabel('Frequency')
ax1.set_title('Working Experience vs. Employment Status (0-30)')

# Customize the x-axis tick labels
ax1.set_xticks(np.arange(len(experience_categories)))
ax1.set_xticklabels(experience_categories, rotation=45)

# Add legend
ax1.legend(title='Status')

# Figure 2: Range 0-10
fig2, ax2 = plt.subplots()

# Smooth each line using a rolling window average with window size 3
smoothed_data = frequencies[:11].rolling(window=3, center=True).mean()
smoothed_data.plot(kind='line', ax=ax2, linestyle='--', linewidth=2)

# Add labels and title to the plot
ax2.set_xlabel('Working Experience(0-10)')
ax2.set_ylabel('Frequency')
ax2.set_title('Working Experience vs. Employment Status (0-10)')

# Customize the x-axis tick labels
ax2.set_xticks(np.arange(11))
ax2.set_xticklabels(experience_categories[:11], rotation=45)

# Add legend
ax2.legend(title='Status')

# Figure 3: Range 11-30
fig3, ax3 = plt.subplots()

# Smooth each line using a rolling window average with window size 3
smoothed_data = frequencies[11:].rolling(window=3, center=True).mean()
smoothed_data.plot(kind='line', ax=ax3, linestyle='--', linewidth=2)

# Add labels and title to the plot
ax3.set_xlabel('Working Experience(11-30)')
ax3.set_ylabel('Frequency')
ax3.set_title('Working Experience vs. Employment Status (11-30)')

# Customize the x-axis tick labels
ax3.set_xticks(np.arange(11, len(experience_categories)))
ax3.set_xticklabels(experience_categories[11:], rotation=45)

# Add legend
ax3.legend(title='Status')

# Display the plots
plt.show()
```
