## Salary Data Pre Processing

```python
pd1 = sl_copy.copy()
```

```python
# Convert each of these values to NaN.
pd1 = pd1.where(pd1 != '\\N')
pd1.Net_salary = pd1.Net_salary.where(pd1.Net_salary != 0)

print(pd1.isnull().sum())
```

```python
pd1 = pd1.dropna(subset=['SiteNo'])
pd1
```

```python
print(pd1.isnull().sum())
```

```python
pd_copy = pd1.copy()
pd1
```

```python
# Keep track of rows that are changed
changed_rows = []

# Iterate over the DataFrame rows
for index, row in pd_copy.iterrows():
    if row['Net_salary'] == 'nan': # If Net_salary is empty
        area = row['Area']
        site_no = row['SiteNo']
        
        # Find rows where Area=X, SiteNo=Y, and Net_salary != 0
        matching_rows = pd_copy[(pd_copy['Area'] == area) & (pd_copy['SiteNo'] == site_no) & (pd_copy['Net_salary'] != 'nan')]
        
        if len(matching_rows) > 0:
            # Calculate the mean of Net_salary for matching rows
            mean_salary = matching_rows['Net_salary'].mean()
            pd_copy.at[index, 'Net_salary'] = mean_salary
            changed_rows.append((index, mean_salary))
        else:
            # Calculate the total mean of Net_salary
            total_mean_salary = pd_copy['Net_salary'].mean()
            pd_copy.at[index, 'Net_salary'] = total_mean_salary
            changed_rows.append((index, total_mean_salary))

# Print the changed rows and their upd_copyated values
for index, value in changed_rows:
    print(f"Row {index} changed. Upd_copyated Net_salary: {value}")

# Preview the upd_copyated DataFrame
print(changed_rows)
```
